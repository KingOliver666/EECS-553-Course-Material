```python
from torchvision import transforms

# Define a data augmentation pipeline
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.9),
    transforms.RandomRotation(degrees=100),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.9, hue=0.1),
    transforms.RandomResizedCrop(size=224, scale=(0.8, 1.0)),
])

# Apply the pipeline to an image
from PIL import Image
image = Image.open('CNN_training_plot.png')
augmented_image = transform(image)
```
